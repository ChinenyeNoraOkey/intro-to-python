#### Question 3

Write a Python program that:
Takes a total number of minutes as an integer (e.g., 547).
Calculates and prints how many hours and remaining minutes there are.


```python
total_minutes = int(input("Enter the total number of minutes: "))
hours = total_minutes // 60
minutes = total_minutes % 60
print(f"{total_minutes} minutes is equal to {hours} hour(s) and {minutes} minute(s).")
```
